```python
import pandas as pd

# Load the CSV files
tabnet = pd.read_csv("Tabnet_submission (2.4682).csv")
yunbase = pd.read_csv("Yunbase submission (2.4697).csv")
lightgbm = pd.read_csv("submission_LightGBM_BAG_L1.csv")

# Define RMSE values for weighted averaging
rmse_tabnet = 2.4682
rmse_yunbase = 2.4697
rmse_lightgbm = 2.5976

# Simple Average
simple_avg = tabnet.copy()
simple_avg['bg+1:00'] = (tabnet['bg+1:00'] + yunbase['bg+1:00'] + lightgbm['bg+1:00']) / 3
simple_avg.to_csv("submission_simple_avg.csv", index=False)

# Weighted Average (inverse RMSE as weights)
weights = [1 / rmse_tabnet, 1 / rmse_yunbase, 1 / rmse_lightgbm]
total_weight = sum(weights)
weighted_avg = tabnet.copy()
weighted_avg['bg+1:00'] = (
    (tabnet['bg+1:00'] * weights[0] + yunbase['bg+1:00'] * weights[1] + lightgbm['bg+1:00'] * weights[2]) 
    / total_weight
)
weighted_avg.to_csv("submission_weighted_avg.csv", index=False)

# Rank Average
tabnet['rank'] = tabnet['bg+1:00'].rank()
yunbase['rank'] = yunbase['bg+1:00'].rank()
lightgbm['rank'] = lightgbm['bg+1:00'].rank()
rank_avg = tabnet.copy()
rank_avg['bg+1:00'] = (tabnet['rank'] + yunbase['rank'] + lightgbm['rank']) / 3
rank_avg['bg+1:00'] = rank_avg['bg+1:00'].rank()  # Map back to original range
rank_avg.to_csv("submission_rank_avg.csv", index=False)

print("Files generated: submission_simple_avg.csv, submission_weighted_avg.csv, submission_rank_avg.csv")
```

```python
# Drop the 'rank' column before saving
rank_avg = rank_avg[['id', 'bg+1:00']]  # Select only the required columns
rank_avg.to_csv("submission_rank_avg.csv", index=False)

print("Corrected submission file generated: submission_rank_avg.csv")
```
