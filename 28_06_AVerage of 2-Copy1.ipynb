```python
import pandas as pd

# Load the CSV files
tabnet = pd.read_csv("Tabnet_submission (2.4682).csv")
yunbase = pd.read_csv("Yunbase submission (2.4697).csv")

# Define RMSE values for weighted averaging
rmse_tabnet = 2.4682
rmse_yunbase = 2.4697

# Simple Average
simple_avg = tabnet.copy()
simple_avg['bg+1:00'] = (tabnet['bg+1:00'] + yunbase['bg+1:00'] + .37) / 3
simple_avg.to_csv("submission_simple_avg_decreased.csv", index=False)

# Weighted Average (inverse RMSE as weights)
weights = [1 / rmse_tabnet, 1 / rmse_yunbase]
total_weight = sum(weights)
weighted_avg = tabnet.copy()
weighted_avg['bg+1:00'] = (
    (tabnet['bg+1:00'] * weights[0] + yunbase['bg+1:00'] * weights[1]) 
    / total_weight
)
weighted_avg.to_csv("submission_weighted_avg.csv", index=False)

print("Files generated: submission_simple_avg.csv, submission_weighted_avg.csv")
```
